# Research-Grade EDA: FakeAVCeleb Dataset Analysis
This notebook provides a deep technical dive into the **FakeAVCeleb v1.2** dataset, focusing on generative methodologies, demographic distributions, and identity swapping matrices to inform biometric robustness evaluations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

os.makedirs('eda_plots', exist_ok=True)
df = pd.read_csv('FakeAVCeleb_v1.2/meta_data.csv')
df.head()
```

## 1. Generative Methodology Analysis (Lip-Sync vs. Face-Swap)
Comparing the prevalence of different synthesis techniques. Dominance of Wav2Lip signifies high-quality lip-audio synchronization.

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='method', data=df, order=df['method'].value_counts().index, palette='magma')
plt.title('Deepfake Generation Methods (Wav2Lip vs. FSGAN)')
plt.tight_layout()
plt.savefig('eda_plots/method_distribution.png')
plt.show()
```

## 2. Subject Density & Dataset Balance
Identifying how many samples each unique source celebrity contributes to the dataset.

```python
subj_counts = df['source'].value_counts()
plt.figure(figsize=(10, 6))
sns.histplot(subj_counts, bins=30, kde=True, color='teal')
plt.title('Dataset Density: Samples per Unique Source Identity')
plt.xlabel('Number of Samples')
plt.tight_layout()
plt.savefig('eda_plots/subject_density.png')
plt.show()
```

## 3. Bivariate Mapping: Method vs. Race
Heatmap analyzing the distribution of generation methods across different physical demographic markers.

```python
pivot_df = df.groupby(['race', 'method']).size().unstack(fill_value=0)
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_df, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Methodology vs. Demographic Race (Heatmap Analysis)')
plt.tight_layout()
plt.savefig('eda_plots/method_race_heatmap.png')
plt.show()
```

## 4. Identity Mapping Matrix (Source vs. Target)
Visualizing the identity swapping variety by mapping the top 10 impersonated sources against their target base videos.

```python
top_sources = df['source'].value_counts().index[:10]
matrix_df = df[df['source'].isin(top_sources)].groupby(['source', 'target1']).size().unstack(fill_value=0)
plt.figure(figsize=(12, 8))
sns.heatmap(matrix_df, cmap='Purples')
plt.title('Source-Target Identity Swapping Matrix (Top 10)')
plt.tight_layout()
plt.savefig('eda_plots/identity_matrix.png')
plt.show()
```
